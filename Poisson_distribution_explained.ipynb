```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import factorial
import pandas as pd
import seaborn as sns
```

A variable is Poisson distributed when the probability of grabbing $k$ samples within a fixed number of samples $n\geq k$ occurs with a constant mean rate, under the assumption that samples are dragged independently

```python
def poisson(end, rt, start=0):
    x = np.linspace(start, end, int((end - start)*10))
    return rt**x * np.exp(-rt) / factorial(x)
```

```python
df1 = pd.Series(poisson(20, 0))
df2 = pd.Series(poisson(20, 5))
df3 = pd.Series(poisson(20, 10))
```
