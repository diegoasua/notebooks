```python
import pandas as pd
import altair as alt
from pathlib import Path
```

```python
source_folder = Path(r"C:\Users\diego\logs")
```

```python
logs = []
for log_name in ["dispatcher",  "scanner", "external_comm", "main", "camera", "saver"]: #"main",
    log = pd.read_csv(source_folder / (log_name + ".txt"), names=["ns","type", "id", "is_sender", "value"])
    log["source"] = log_name
    logs.append(log)

all_logs = pd.concat(logs)

all_logs = all_logs.sort_values("ns")

all_logs["time (s)"] = (all_logs.ns - all_logs.ns.min())/1e9
```

```python
selection = alt.selection_interval(bind='scales')

alt.Chart(all_logs).mark_tick().encode(
    x="time (s)",
    y="source",
    tooltip="id",
    color="type",
).add_selection(selection)
```
